```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')
```

```python
def rk4(t, y0, dt, n, f):
    k1 = f(t, y0)
    
    k2 = f(t+dt/2, y0 + k1*dt/2)
    
    k3 = f(t+dt/2, y0 + k2*dt/2)
    
    k4 = f(t+dt, y0 + k3*dt)
    y = y0 + (k1 + 2*k2 + 2*k3 + k4)*dt/6
    
    return y
```

```python
def f(t, y):
    # y in R^n
    return -y
```

```python
N  = 100; a = 0; b =10; h = (b-a)/N; t = 0; y = 10;
result = np.array([]); tt = np.array([]);
while (t < b):
    if t+h > b:
        h = b-t
        
    y = rk4(t, y, h, 1, f)
    result = np.append(result, y)
    tt = np.append(tt, t)
    t = t + h
fig, ax = plt.subplots()
_ = ax.plot(tt,result)
```

```python
def Lorenz(t, x, p=10, r = 28, b =8/3):
    # x in R^3
    return np.array([
        -p*x[0] + p*x[1],
        -x[0]*x[2] + r*x[0] - x[1],
        x[0]*x[1] - b*x[2]
    ])
```

```python
from mpl_toolkits.mplot3d import Axes3D

N  = 10000; a = 0; b =100; h = (b-a)/N; t = 0;

x = np.array([1,1,1])
result = np.zeros((N,3))
result[0] = x[:]

for i in range(N):
    t = i*h
    x = rk4(t, x, h, 3, Lorenz)
    result[i] = x

t = np.linspace(0,100, N)
fig = plt.figure(figsize=(5,5))
# for i in range(3):
#     ax = plt.subplot(3,1,i+1)
#     ax.plot(t,result[:, i])
ax = Axes3D(fig)
_= ax.plot(result[:,0],result[:,1],result[:,2])
```

```python
def f_osi(t,y):
    return np.array([y[1], -y[0]])
```

```python
N = 1000; a = 0; b = 100; h = (b-a)/N;

y = np.array([1,0])
result = np.zeros((N,2))
result[0] = y[:]

for i in range(N):
    t = i*h
    y = rk4(t,y,h,2,f_osi)
    result[i] = y
    
E_k = result[:,1]**2

tt = np.linspace(a,b,N)
exact_x = [np.cos(t) for t in tt]
exact_v= [-np.sin(t) for t in tt]
eps = np.abs(result[:,0]-exact_x)

fig, ax = plt.subplots()
ax.plot(tt,result[:,0],label="numeric x")
ax.plot(tt,exact_x,label="exact x")
ax.plot(tt,eps,label="eps")
# plt.plot(tt,E_k,label="Kinetic Energy")
ax.legend()
```

```python
a = 0; b = 100; 

k1 = 2; k2=5;
ks = np.arange(k1,k2)
eps = np.zeros(len(ks))

for k in range(k1,k2):
    N = 10**k
    h = (b-a)/N;
    y = np.array([1,0])
    result = np.zeros((N,2))
    result[0] = y[:]
    for i in range(N):
        t = i*h
        y = rk4(t,y,h,2,f_osi)
        result[i] = y
    tt = np.linspace(a,b,N)
    exact_x = [np.cos(t) for t in tt]
    eps[k-k1] = np.abs(exact_x-result[:,0]).max()


hs = np.array([(b-a)/10**k for k in ks])

fig, ax = plt.subplots()
ax.set_xlabel("h")
ax.set_ylabel("eps")
ax.plot(hs,eps,'o-')
```
