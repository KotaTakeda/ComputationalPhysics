```python
import numpy as np
import matplotlib.pyplot as plt
import time
N = 1000
```

```python
a = np.arange(N**2).reshape(N,N)
b = np.arange(N**2).reshape(N,N)

sum = np.zeros(3)
```

列を先に変化、早い

```python
t = np.zeros(N)
t1 = time.time()
for i in range(1,N):
    for j in range(1,N):
        sum[1] += b[i,j]
        if i==1:
            t[j] = time.time() - t[j-1]
t2 = time.time()
print(t2-t1)
plt.plot(t)
```

行を先に変化、遅い

```python
t = np.zeros(N)
t1 = time.time()
for j in range(1,N):
    for i in range(1,N):
        sum[0] += a[i,j]
        if j==1:
            t[i] = time.time() - t[i-1]
t2  =time.time()
print(t2-t1)
plt.plot(t)
```

```python
x = np.zeros(N)
t1 = time.time()
for i in range(N):
    x[i] = a[i,1]
t2 = time.time()
for j in range(N):
    x[j] = a[1,j]
t3 = time.time()

print("i ",t2-t1,"\nj ",t3-t2)
```
